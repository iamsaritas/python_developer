# Simple Data Manipulation with Pandas
# 1. Learn how to read CSV files and manipulate data frames using Pandas:

```python
import pandas as pd
# Read CSV file into a DataFrame
df = pd.read_csv('Data _Cleaning_ and_ Preprocessing.csv')

data = pd.read_csv("Data _Cleaning_ and_ Preprocessing.csv")
print(data)
data.head()
```

#2- Perform simple data cleaning tasks such as handling missing values and removing duplicates:

```python
# Handling missing values
df_cleaned = df.dropna()  # Drop rows with any missing values

data = pd.read_csv("Data _Cleaning_ and_ Preprocessing.csv")
print(data)
data.head()
```

```python
# Removing duplicates
df_cleaned = df_cleaned.drop_duplicates()

print("DataFrame after handling missing values and removing duplicates:")
print(df_cleaned)
print("\n")

data = pd.read_csv("Data _Cleaning_ and_ Preprocessing.csv")
print(data)
data.head()
```

# 3- Practice basic data manipulation operations like filtering, sorting, and grouping data:

```python
# Filtering rows where 'Y-Kappa' is greater than 25
filtered_df = df[df['Y-Kappa'] > 25]
print("Filtered DataFrame where Y-Kappa > 25:")
print(filtered_df)
print()

data = pd.read_csv("Data _Cleaning_ and_ Preprocessing.csv")
print(data)
data.head()
```

```python
# Sorting DataFrame by 'ChipRate' column in descending order
sorted_df = df.sort_values(by='ChipRate', ascending=False)
print("Sorted DataFrame by ChipRate in descending order:")
print(sorted_df)
print()

data = pd.read_csv("Data _Cleaning_ and_ Preprocessing.csv")
print(data)
data.head()
```

```python
# Grouping by 'Observation' and calculating the mean for each group
grouped_df = df.groupby('Observation').mean()
print("Grouped DataFrame by Observation and mean calculated for each group:")
print(grouped_df)

data = pd.read_csv("Data _Cleaning_ and_ Preprocessing.csv")
print(data)
data.head()
```
